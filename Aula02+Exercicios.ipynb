###Desafio 03: Plotar os histogramas com seaborn

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data =dados.loc[:,'g_0':'g_771'].describe().T['mean'], bins = 30)
ax.set_title('MÃ©dia da variaÃ§Ã£o do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('MÃ©dia', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['min'], bins = 30)
ax.set_title('Minimo da variaÃ§Ã£o do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Minimo', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['max'], bins = 30)
ax.set_title('MÃ¡ximo da variaÃ§Ã£o do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('MÃ¡ximo', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(dados.loc[:, 'c_0': 'c_99'].describe().T['mean'], bins = 50)
ax.set_title('MÃ©dia da variaÃ§Ã£o do RNAs dos Compostos Celulares', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('MÃ©dia', size = 15)
```

###Desafio 04: Estudar sobre as estatÃ­sticas retornadas no .describe

Count: Retorna o nÃºmero de linhas que aquela coluna possui

Mean: Retorna a mÃ©dia aritmetica. Deve ser usada com cuidado, pois, se utilizarmos varios itens com valores baixos e um com valores altos, a media serÃ¡ um valor mediano, o que pode nÃ£o ser o resultado esperado

std: Retorna o desvio de padrÃ£o, ou seja, a maior variaÃ§Ã£o presente da mÃ©dia

min e max: retornam, sucessivamente, o valor minimo e o maximo dos dados

25%, 50%, 70%: retornam, sucessivamente, 25, 50 e 75% da quantidade dos dados. Eles tambÃ©m podem ser chamados de "Quartils", e o 50% tambÃ©m pode ser chamado de "mediana"

###Desafio 05: Refletir sobre a manipulaÃ§Ã£o do tamanho das visualizaÃ§Ãµes.

Acredito que a sensaÃ§Ã£o de "distorÃ§Ã£o" seja pois, com a quantidade tanto de dados quanto de excedentes, a Boxplot acabe ficando muito achatada, por isso o tamanho de sua visualizaÃ§Ã£o Ã© um dos passos necessarios para que se tenha uma visÃ£o mais limpa.

###Desafio 06: Fazer outras analises com o boxplot e atÃ© com o histograma.

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['std'], bins = 30)
ax.set_title('Desvio de PadrÃ£o(STD) dos RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('STD', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(dados.loc[:, 'c_0': 'c_99'].describe().T['std'], bins = 50)
ax.set_title('Desvio de PadrÃ£o(STD) dos RNAs dos Compostos Celulares', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('MÃ©dia', size = 15)
```
