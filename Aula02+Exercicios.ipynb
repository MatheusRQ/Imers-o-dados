###Desafio 03: Plotar os histogramas com seaborn

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data =dados.loc[:,'g_0':'g_771'].describe().T['mean'], bins = 30)
ax.set_title('Média da variação do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Média', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['min'], bins = 30)
ax.set_title('Minimo da variação do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Minimo', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['max'], bins = 30)
ax.set_title('Máximo da variação do RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Máximo', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(dados.loc[:, 'c_0': 'c_99'].describe().T['mean'], bins = 50)
ax.set_title('Média da variação do RNAs dos Compostos Celulares', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Média', size = 15)
```

###Desafio 04: Estudar sobre as estatísticas retornadas no .describe

Count: Retorna o número de linhas que aquela coluna possui

Mean: Retorna a média aritmetica. Deve ser usada com cuidado, pois, se utilizarmos varios itens com valores baixos e um com valores altos, a media será um valor mediano, o que pode não ser o resultado esperado

std: Retorna o desvio de padrão, ou seja, a maior variação presente da média

min e max: retornam, sucessivamente, o valor minimo e o maximo dos dados

25%, 50%, 70%: retornam, sucessivamente, 25, 50 e 75% da quantidade dos dados. Eles também podem ser chamados de "Quartils", e o 50% também pode ser chamado de "mediana"

###Desafio 05: Refletir sobre a manipulação do tamanho das visualizações.

Acredito que a sensação de "distorção" seja pois, com a quantidade tanto de dados quanto de excedentes, a Boxplot acabe ficando muito achatada, por isso o tamanho de sua visualização é um dos passos necessarios para que se tenha uma visão mais limpa.

###Desafio 06: Fazer outras analises com o boxplot e até com o histograma.

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados.loc[:,'g_0':'g_771'].describe().T['std'], bins = 30)
ax.set_title('Desvio de Padrão(STD) dos RNAs dos Genes', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('STD', size = 15)
```

```python
plt.figure(figsize = (10,8))
ax = sns.histplot(dados.loc[:, 'c_0': 'c_99'].describe().T['std'], bins = 50)
ax.set_title('Desvio de Padrão(STD) dos RNAs dos Compostos Celulares', size = 20)
plt.ylabel('Quantidade', size = 15)
plt.xlabel('Média', size = 15)
```
